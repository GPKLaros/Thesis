```python
import pandas as pd
import numpy as np
##Loading the complete fifa database
complete_data = pd.read_csv('complete_data_test.csv', sep=',')
for i in range(len(complete_data)):
    if complete_data['year'][i] == 2021:
        complete_data['contract_valid_until'][i] = complete_data['club_contract_valid_until'][i]
```

```python
#Removing variables that are unnecessary and renaming columns.
complete_data.drop(['Unnamed: 0_x', 'Unnamed: 0', 'player_name', 'fee', 'transfer_movement', 'club_name_y', 'league_name_y', 'league_rank', 'season', 'player_url', 'id-name', 'Unnamed: 0_y', 'sofifa_id', 'short_name', 'long_name', 'age_y', 'dob', 'player_positions', 'work_rate', 'body_type', 'real_face', 'player_tags', 'player_traits', 'team_position', 'team_jersey_number', 'loaned_from', 'joined', 'nation_position', 'nation_jersey_number', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_reflexes', 'gk_speed', 'gk_positioning', 'mentality_composure', 'defending_marking', 'value_eur', 'wage_eur'], axis = 1, inplace = True)
complete_data.rename(columns = {'club_name_x':'buying_club', 'age_x':'age','club_involved_name':'selling_club', 'league_name_x':'buying_league'}, inplace = True)
complete_data.drop(complete_data.iloc[:, 56:], inplace = True, axis = 1)
```

```python
#Turning position into position_group
defenders = ['Centre-Back', 'Left-Back', 'Right-Back']
midfielders = ['Attacking Midfield', 'Defensive Midfield', 'Central Midfield', 'Right Midfield', 'Left Midfield']
attackers = ['Centre-Forward', 'Left Winger', 'Right Winger', 'Second Striker']
complete_data['position_group'] = 0
for i in range(len(complete_data)):
    if complete_data['position'][i] in defenders:
        complete_data['position_group'][i] = 'Defender'
    elif complete_data['position'][i] in midfielders:
        complete_data['position_group'][i] = 'Midfielder'
    elif complete_data['position'][i] in attackers:
        complete_data['position_group'][i] = 'Attacker'
complete_data = complete_data[complete_data["position_group"].str.contains('Midfielder') == True]
```

```python
#Variables changed to categoric variables
complete_data['buying_club'] = pd.Categorical(complete_data['buying_club'])
complete_data['selling_club'] = pd.Categorical(complete_data['selling_club'])
complete_data['transfer_period'] = pd.Categorical(complete_data['transfer_period'])
complete_data['buying_league'] = pd.Categorical(complete_data['buying_league'])
complete_data['nationality'] = pd.Categorical(complete_data['nationality'])
complete_data['international_reputation'] = pd.Categorical(complete_data['international_reputation'])
complete_data['position'] = pd.Categorical(complete_data['position'])
complete_data['weak_foot'] = pd.Categorical(complete_data['weak_foot'])
complete_data['skill_moves'] = pd.Categorical(complete_data['skill_moves'])
complete_data['international_reputation'] = pd.Categorical(complete_data['international_reputation'])
complete_data['position_group'] = pd.Categorical(complete_data['position_group'])
```

```python
#Taking all goalkeepers out of the data
data_without_gk = complete_data[complete_data["position"].str.contains('Goalkeeper') == False]
data_with_gk = complete_data[complete_data["position"].str.contains('Goalkeeper') == True]
data_without_gk = data_without_gk.reset_index(drop=True)
data_with_gk = data_with_gk.reset_index(drop=True)
```

```python
#Creating the dummy variables
position_group_dummy = pd.get_dummies(data_without_gk['position_group'], prefix='position')
#2 in iloc because otherwise it would take 0 (for goalkeepers) as the base category, because of the 2 it is attacker
position_group_dummy = position_group_dummy.iloc[: , 2:]
buying_league_dummy = pd.get_dummies(data_without_gk['buying_league'], prefix='buying_league')
buying_league_dummy = buying_league_dummy.iloc[: , 1:]
nationality_dummy = pd.get_dummies(data_without_gk['nationality'], prefix='nationality')
nationality_dummy = nationality_dummy.iloc[: , 1:]
preferred_foot_dummy = pd.get_dummies(data_without_gk['preferred_foot'], prefix='preferred_foot')
preferred_foot_dummy = preferred_foot_dummy.iloc[: , 1:]
transfer_period_dummy = pd.get_dummies(data_without_gk['transfer_period'], prefix='transfer_period')
transfer_period_dummy = transfer_period_dummy.iloc[: , 1:]

#Dropping the initial values of the new dummies and the goalkeeper skills
temp = data_without_gk.drop(['position', 'release_clause_eur', 'buying_league', 'selling_club', 'buying_club', 'nationality', 'preferred_foot', 'transfer_period', 'position_group'], axis = 1)
model_data = pd.concat([temp, position_group_dummy, buying_league_dummy, nationality_dummy, preferred_foot_dummy, transfer_period_dummy], axis = 1)
```

```python
#Number of rows with NA
model_data.isnull().any(axis=1).sum()
```

```python
#Number of empty cells per column
for col in model_data:
    print(col, ':', (model_data[col].isna().sum()/len(model_data))*100)
```
