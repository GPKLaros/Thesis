```python
import pandas as pd
import numpy as np
##Loading the complete fifa database
all_fifa_data = pd.read_csv('all_fifa_data.csv')
#Loading the complete transfer database
#transfer_data_manual.csv is the data with player_url for the players with a single name put in manually
all_transfer_data = pd.read_csv('transfer_data_manual.csv', sep=';')
#Merging the two datasets on player_url
complete_data = pd.merge(all_transfer_data, all_fifa_data, on = 'player_url')
complete_data
```

```python
##Some exploratory code
#Finding the transfer with the highest transfer fee
complete_data.loc[complete_data['fee_cleaned'] == complete_data['fee_cleaned'].max()]
#Finding the transfer(s) with the highest transfer fee
complete_data.loc[complete_data['fee_cleaned'] == complete_data['fee_cleaned'].min()]
#Player with the highest rating that made a transfer
complete_data.loc[complete_data['overall'] == complete_data['overall'].max()]
#Player with the highest potential that made a transfer
complete_data.loc[complete_data['potential'] == complete_data['potential'].max()]
```

```python
#putting nationality_name on the place of nationality for players that do not have a 
complete_data.loc[complete_data['nationality'].isnull(),'nationality'] = complete_data['nationality_name']
```

```python
complete_data.to_csv('complete_data_test.csv')
```
