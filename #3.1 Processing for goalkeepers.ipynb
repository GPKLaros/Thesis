```python
import pandas as pd
import numpy as np
##Loading the complete fifa database
complete_data = pd.read_csv('complete_data_test.csv', sep=',')
for i in range(len(complete_data)):
    if complete_data['year'][i] == 2021:
        complete_data['contract_valid_until'][i] = complete_data['club_contract_valid_until'][i]
```

```python
data_with_gk = complete_data[complete_data["position"].str.contains('Goalkeeper') == True]
data_with_gk = data_with_gk.reset_index(drop=True)
```

```python
#In FIFA from 2021, the goalkeeper variables are renamed and therefore for FIFA from 2021 we put the information of goalkeeper skills in the variable in the old variables of goalkeeper skills
for i in range(len(data_with_gk)):
    if data_with_gk['year'][i] == 2021:
        data_with_gk['gk_diving'][i] = data_with_gk['goalkeeping_diving'][i]
        data_with_gk['gk_handling'][i] = data_with_gk['goalkeeping_handling'][i]
        data_with_gk['gk_positioning'][i] = data_with_gk['goalkeeping_positioning'][i]
        data_with_gk['gk_reflexes'][i] = data_with_gk['goalkeeping_reflexes'][i]
        data_with_gk['gk_kicking'][i] = data_with_gk['goalkeeping_kicking'][i]
        data_with_gk['gk_speed'][i] = data_with_gk['goalkeeping_speed'][i]
```

```python
#Removing variables that are unnecessary and renaming columns.
data_with_gk.drop(['Unnamed: 0_x', 'Unnamed: 0', 'player_name', 'fee', 'transfer_movement', 'club_name_y', 'league_name_y', 'league_rank', 'season', 'player_url', 'id-name', 'Unnamed: 0_y', 'sofifa_id', 'short_name', 'long_name', 'age_y', 'dob', 'player_positions', 'work_rate', 'body_type', 'real_face', 'player_tags', 'team_position', 'team_jersey_number', 'loaned_from', 'joined', 'nation_position', 'nation_jersey_number', 'pace', 'dribbling', 'shooting', 'passing', 'physic', 'defending', 'value_eur', 'wage_eur', 'release_clause_eur'], axis = 1, inplace = True)
data_with_gk.rename(columns = {'club_name_x':'buying_club', 'age_x':'age','club_involved_name':'selling_club', 'league_name_x':'buying_league'}, inplace = True)
data_with_gk.drop(complete_data.iloc[:, 91:], inplace = True, axis = 1)
```

```python
#Variables that are categoric
data_with_gk['buying_club'] = pd.Categorical(data_with_gk['buying_club'])
data_with_gk['selling_club'] = pd.Categorical(data_with_gk['selling_club'])
data_with_gk['transfer_period'] = pd.Categorical(data_with_gk['transfer_period'])
data_with_gk['buying_league'] = pd.Categorical(data_with_gk['buying_league'])
data_with_gk['nationality'] = pd.Categorical(data_with_gk['nationality'])
data_with_gk['international_reputation'] = pd.Categorical(data_with_gk['international_reputation'])
data_with_gk['position'] = pd.Categorical(data_with_gk['position'])
data_with_gk['weak_foot'] = pd.Categorical(data_with_gk['weak_foot'])
data_with_gk['skill_moves'] = pd.Categorical(data_with_gk['skill_moves'])
data_with_gk['international_reputation'] = pd.Categorical(data_with_gk['international_reputation'])
```

```python
#creating dummies for the variables that need it
buying_league_dummy = pd.get_dummies(data_with_gk['buying_league'], prefix='buying_league')
buying_league_dummy = buying_league_dummy.iloc[: , 1:]
nationality_dummy = pd.get_dummies(data_with_gk['nationality'], prefix='nationality')
nationality_dummy = nationality_dummy.iloc[: , 1:]
preferred_foot_dummy = pd.get_dummies(data_with_gk['preferred_foot'], prefix='preferred_foot')
preferred_foot_dummy = preferred_foot_dummy.iloc[: , 1:]
transfer_period_dummy = pd.get_dummies(data_with_gk['transfer_period'], prefix='transfer_period')
transfer_period_dummy = transfer_period_dummy.iloc[: , 1:]

#Dropping the initial values of the new dummies and the goalkeeper skills
temp = data_with_gk.drop(['position', 'buying_league', 'selling_club', 'buying_club', 'nationality', 'preferred_foot', 'transfer_period', 'player_traits', 'mentality_composure', 'defending_marking'], axis = 1)
data_with_gk = pd.concat([temp, buying_league_dummy, nationality_dummy, preferred_foot_dummy, transfer_period_dummy], axis = 1)
```

```python
#Number of rows with NA
data_with_gk.isnull().any(axis=1).sum()
```

```python
#Number of empty cells per column
for col in data_with_gk:
    print(col, ':', (data_with_gk[col].isna().sum()/len(data_with_gk))*100)
```

```python
#Dropping the rows with an empty value
data_with_gk = data_with_gk.dropna()
data_with_gk = data_with_gk.reset_index(drop=True)
data_with_gk.to_csv('goalkeeper_data.csv', index = False)
data_with_gk
```
